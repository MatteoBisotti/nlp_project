```python
import pandas as pd
from datasets import load_dataset
eu_debates_dataset = load_dataset('coastalcph/eu_debates', split='train', trust_remote_code=True)
dataset = eu_debates_dataset.to_pandas()
dataset
```

```python
dataset = dataset[dataset['text'].str.split().apply(len) < 370] # take only speeches with length shorter than 370 
dataset = dataset[dataset['translated_text'].isnull()]          # take only speeches in english
dataset = dataset.drop(['translated_text'], axis=1)
dataset
```

```python
dataset.to_csv("../data/politics_texts.csv", index=False)
```
