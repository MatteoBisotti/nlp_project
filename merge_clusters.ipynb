```python
import pandas as pd
```

```python
dataset = pd.read_csv('data/final_dataset.csv')
dataset_clusters = pd.read_csv('data/subreddit_embeddings.csv')
```

```python
merge_dataset = dataset.merge(dataset_clusters[['subreddit', 'cluster']], on='subreddit', how='left')
```

```python
merge_dataset
```

```python
merge_dataset['cluster'].value_counts().sort_index()
```

| 6       | Videogames       |

```python
# mapping cluster -> label
cluster_to_topic = {
    0: 'humor',
    1: 'other',
    2: 'sport',
    3: 'politics',
    4: 'love and relationship',
    5: 'film and TV series',
    6: 'videogames'
}

merge_dataset['topic'] = merge_dataset['cluster'].map(cluster_to_topic)
```

```python
merge_dataset
```

```python
merge_dataset.to_csv('data/dataset_topics.csv', index=False)
```
